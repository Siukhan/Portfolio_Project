### PROBLEM EXPLANATION

Suppose a person takes a loan from a bank.

The bank charges interest and asks the borrower to pay a fixed amount every month, called EMI-Equated Monthly Installment.

Our goal is to calculate:

    1. Monthly EMI

    2. Total amount paid

    3. Total interest paid

### MATHEMATICAL FORMULA


$$
EMI = \frac{P \times r \times (1 + r)^n}{(1 + r)^n - 1}
$$

This formula is derived from **compound interest**, which is why banks earn more over time.‚Äù

### PYTHON CODE
üìå EMI CALCULATOR WITH GRAPH

```python
# Import necessary libraries
import matplotlib.pyplot as plt

# Input values
principal = float(input("Enter loan amount: "))
annual_rate = float(input("Enter annual interest rate (%): "))
years = int(input("Enter loan duration (years): "))

# Conversions
monthly_rate = annual_rate / (12 * 100)
months = years * 12

# EMI formula
emi = (principal * monthly_rate * (1 + monthly_rate) ** months) / \
      ((1 + monthly_rate) ** months - 1)

total_payment = emi * months
total_interest = total_payment - principal

print("\n===== Loan Summary =====")
print(f"Monthly EMI: Rs. {emi:.2f}")
print(f"Total Payment: Rs. {total_payment:.2f}")
print(f"Total Interest: Rs. {total_interest:.2f}")

# Graph: Loan repayment
months_list = list(range(1, months + 1))
emi_list = [emi] * months_list.__len__()

plt.plot(months_list, emi_list)
plt.xlabel("Months")
plt.ylabel("Monthly EMI (Rs)")
plt.title("Loan EMI Over Time")
plt.show()
```
